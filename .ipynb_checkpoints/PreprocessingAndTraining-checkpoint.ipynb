# 1 - Data preprocessing

```python
import xgboost as xgb
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import joblib
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, confusion_matrix, recall_score, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_sample_weight, compute_class_weight
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from collections import Counter
from xgboost import plot_importance

data = pd.read_csv('Dataset/Data.csv')
label = pd.read_csv('Dataset/Label.csv')

# Each row of data has a label to identify which type of attack it is, so we join eachother
joined = pd.concat([data, label], axis=1)

# For possible infinites, we transform them into NaN (nulls), and errase the rows affected
joined.replace([np.inf, -np.inf], np.nan, inplace=True)
after_drop = len(joined)
joined.dropna(inplace=True)
print(f"NaN's found: {after_drop - len(joined)}")

# We should analyze our working set before deciding how to continue
print(" ")
print("Let's analyze the samples we have")
label_counts = joined['Label'].value_counts()
for label, count in label_counts.items():
    percentage = (count / len(joined)) * 100
    print(f"{label}: {count:,} samples ({percentage:.2f}%)")

mapping = {0: 'Normal/Benign', 1: 'Analysis', 2: 'Backdoor', 3: 'DoS', 4: 'Exploits', 5: 'Fuzzers', 6: 'Generic', 7: 'Reconnaissance', 8: 'Shellcode', 9: 'Worms'}
tick_labels = [mapping.get(i, i) for i in label_counts.index]
plt.figure(figsize=(12, 7))
sns.barplot(x=tick_labels, y=label_counts.values, palette='viridis', hue=label_counts.index, legend=False)
plt.title('Attack type distribution')
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Showing a heatmap of the correlation between each feature can help visualizing what we need to do, as well as the features themselves
plt.figure(figsize=(15, 10))
numeric_df = joined.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', xticklabels=True, yticklabels=True)
plt.title('Correlation Heatmap')
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.show()

# The correlation heatmap shows that there are unrelated features, which are not useful for training an AI model, so we erase them
unrel = [col for col in joined.columns if joined[col].nunique() <= 1]
print(f"Non-useful columns (constant value): {unrel}")
joined.drop(columns=unrel, inplace=True)

# Now our heatmap does not present any white, empty lines
plt.figure(figsize=(15, 10))
numeric_df = joined.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', xticklabels=True, yticklabels=True)
plt.title('Correlation Heatmap (Updated)')
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.show()
```

```python
# We have to be sure that no column is something different than a number (if so, we can skip the encoding step)
print(joined.dtypes)

# Now it is imperative that we split our working set into two (label (Y) and the rest of the data columns (X))
# If not, the Neural Network will have the solutions available, and will not learn from the useful data
Y = joined['Label']
X = joined.drop(columns=['Label'])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# Neural Networks require normalized/scaled data to converge properly
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# 2 - Neural Network

## 2.1 Training

```python
print("Preparing Level 1 (Binary Classification)...")
# We transform our labels: 0 remains 0 (Normal), anything > 0 becomes 1 (Attack)
Y_train_bin = (Y_train > 0).astype(int)
Y_test_bin = (Y_test > 0).astype(int)

binary_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(Y_train_bin),
    y=Y_train_bin
)
dict_weights_nn = {0: binary_weights[0], 1: binary_weights[1]}
print(f"NN applied weights -> Normal: {dict_weights_nn[0]:.2f}, Attack: {dict_weights_nn[1]:.2f}")

nn_model = Sequential([
    Input(shape=(X_train_scaled.shape[1],)),
    
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.2), # Prevents overfitting
    
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    Dense(32, activation='relu'),
    BatchNormalization(),
    
    Dense(1, activation='sigmoid') # Sigmoid for binary output (probability between 0 and 1)
])

nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Reducing the learning rate so it can keep learning when the process does not improve
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=0.00005)

# Early stopping to halt training when the validation loss stops improving
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

print("Training Neural Network...")
nn_model.fit(
    X_train_scaled, Y_train_bin, 
    epochs=50, 
    batch_size=256, 
    validation_split=0.1, 
    callbacks=[early_stop, reduce_lr],
    verbose=1
)
```

## 2.2 Optimal threshold search

```python
# We search for the optimal threshold on X_test
# In this case, we need to focus on optimizing the recall (because that helps minimizing the Attacks that get dettected as Normal)
nn_probs_val = nn_model.predict(X_test_scaled).flatten()
Y_test_bin = (Y_test > 0).astype(int)

thresholds = np.arange(0.05, 0.95, 0.01)
best_threshold = 0.5
best_recall = 0

for t in thresholds:
    preds = (nn_probs_val > t).astype(int)
    recall = recall_score(Y_test_bin, preds)
    if recall > best_recall:
        best_recall = recall
        best_threshold = t

print(f"Optimal threshold: {best_threshold:.2f}")
print(f"Attack recall: {best_recall:.4f} ({best_recall*100:.2f}% of detected attacks)")

# How many Normal entries are rejected in our try to minimize attacks passing through
preds_opt = (nn_probs_val > best_threshold).astype(int)
fp = ((preds_opt == 1) & (Y_test_bin == 0)).sum()
total_normal = (Y_test_bin == 0).sum()
print(f"{fp:,} normal traffic labled as attacks ({100*fp/total_normal:.2f}% of fake positives)")
```

## 2.3 Evaluation

```python
print("=== Level 1 Evaluation: Neural Network (Binary) ===\n")

nn_probs_eval = nn_model.predict(X_test_scaled).flatten()
Y_test_bin_eval = (Y_test > 0).astype(int)
nn_preds_eval = (nn_probs_eval > best_threshold).astype(int)

print("Classification Report:")
print(classification_report(Y_test_bin_eval, nn_preds_eval, target_names=["Normal", "Attack"]))

# AUC-ROC
auc = roc_auc_score(Y_test_bin_eval, nn_probs_eval)
print(f"AUC-ROC: {auc:.4f}")

missed_attacks = ((nn_preds_eval == 0) & (Y_test_bin_eval == 1)).sum()
total_attacks = Y_test_bin_eval.sum()

#Confusion matrix
plt.figure(figsize=(6, 4))
cm_bin = confusion_matrix(Y_test_bin_eval, nn_preds_eval)
sns.heatmap(cm_bin, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Normal", "Attack"],
            yticklabels=["Normal", "Attack"])
plt.title('Confusion Matrix - Neural Network')
plt.ylabel('Real')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

# 3 - Attack Identification

## 3.1 XGBoost training

```python
print("\nPreparing Level 2 (Multiclass Attack Classification)...")

# The balanced weights are too agressive, so we apply the squared root factor (balanced = 1, sqr = 0.5)
def get_smoothed_sample_weights(y, smoothing_factor=0.5):
    counter = Counter(y)
    max_count = max(counter.values())
    
    # Smoothed weight for every class
    class_weights = {cls: (max_count / count) ** smoothing_factor for cls, count in counter.items()}
    sample_weights = np.array([class_weights[cls] for cls in y])
    return sample_weights

# We only select the rows from the training set that are attacks
attack_indices_train = Y_train > 0
X_train_attacks = X_train_scaled[attack_indices_train]

# XGBoost expects class labels to start from 0 up to num_classes-1.
# Since we removed 0 (Normal), our attacks are 1 to 9. We subtract 1 to make them 0 to 8.
Y_train_attacks = Y_train[attack_indices_train] - 1

smoothed_weights = get_smoothed_sample_weights(Y_train_attacks, smoothing_factor=0.5)

print("Training XGBoost...")
xgb_model = xgb.XGBClassifier(
    n_estimators=400,
    learning_rate=0.025,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='multi:softprob',
    eval_metric='mlogloss',
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(
    X_train_attacks, 
    Y_train_attacks, 
    sample_weight=smoothed_weights
)

print("\n--- Training Complete! ---")
```

## 3.2 Full Pipeline Evaluation

```python
print("Executing the Neural Network model...")

nn_probs = nn_model.predict(X_test_scaled)
nn_preds = (nn_probs > best_threshold).astype(int).flatten()

final_predictions = np.zeros(len(X_test_scaled), dtype=int)

# We save all the attacks detected
attack_indices_test = np.where(nn_preds == 1)[0]
print(f"NN detected {len(attack_indices_test)} possible attacks from {len(X_test_scaled)} entries.")

if len(attack_indices_test) > 0:
    print("Executing the XGBoost model...")
    
    # Only the ones selected as attacks by the NN
    X_test_attacks = X_test_scaled[attack_indices_test]
    
    xgb_preds = xgb_model.predict(X_test_attacks)
    
    # We map them back to the original label numbers
    final_predictions[attack_indices_test] = xgb_preds + 1

print("\n--- METRICS ---")

target_names = [mapping.get(i, f"Class {i}") for i in sorted(Y.unique())]

print("\nClassification Report:")
print(classification_report(Y_test, final_predictions, target_names=target_names))

# Confusion Matrix
plt.figure(figsize=(12, 8))
cm = confusion_matrix(Y_test, final_predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix - Full Model:')
plt.xlabel('Model Prediction')
plt.ylabel('Real Label')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
